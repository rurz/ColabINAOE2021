```julia
using ColabINAOE2021
using PyPlot
using SparseArrays
using LinearAlgebra
using LaTeXStrings
using NumericalIntegration
```

```julia
# Global Constants

k = 1
N = 32
;
```

```julia
# Initial amplitude vector constructors

pint(q) = sparsevec(Dict(q + 1 => 1), N + 1); # Single amplitude point at 0 ≤ q ≤ N
```

```julia
# Initial amplitude vector

q0 = normalize(Array(pint(0)));
```

```julia
# Time list

t_list = range(0, stop = 60, length = 200);
```

### 1. Circular array

```julia
exc01 = wijct(1, 0, 1, t_list, N, q0);
exc016 = wijct(1, 0, 16, t_list, N, q0);
exc031 = wijct(1, 0, 31, t_list, N, q0);
```

```julia
qc0 = [qc(1, 0, t_list[i], N, q0) for i in 1:length(t_list)];
qc1 = [qc(1, 1, t_list[i], N, q0) for i in 1:length(t_list)];
qc16 = [qc(1, 16, t_list[i], N, q0) for i in 1:length(t_list)];
qc31 = [qc(1, 31, t_list[i], N, q0) for i in 1:length(t_list)];
```

```julia
ET01 = integrate(t_list, exc01, SimpsonEven())
ET016 = integrate(t_list, exc016, SimpsonEven())
ET031 = integrate(t_list, exc031, SimpsonEven())

[ET01, ET016, ET031]
```

```julia
fig_qc = figure(figsize = (10,5))
ax_c = gca()

    ax_c.plot(qc0, "r--", label = L"q_{0}(t)")
    ax_c.plot(qc1, "b-.", label = L"q_{1}(t)")
    ax_c.plot(qc16, "g:", label = L"q_{16}(t)")
    ax_c.plot(qc31, "y-", label = L"q_{31}(t)")

ax_c.set_xlabel(L"\mathrm{Time}\quad [s]")
ax_c.set_ylabel(L"\mathrm{Amplitude}\quad q_{n}(t)")
ax_c.set_xticks([0,50,100,150,200])
ax_c.set_xticklabels([0,15,30,45,60])

ax_c.annotate(L"q_{0}(t)", xy = (1,1))
ax_c.annotate(L"q_{1}(t)", xy = (5,0.5))
ax_c.annotate(L"q_{16}(t)", xy = (55,0.41))
ax_c.annotate(L"q_{31}(t)", xy = (9,0.4))

tight_layout()

legends = ax_c.legend(loc="upper center", ncol = 2, shadow=true, fontsize="12", framealpha = 0.75)

#savefig("circAmp.png", dpi = 300, transparent = true)
```

```julia
fig1 = figure(figsize = (10,5))
ax1 = gca()
    ax1.plot(exc01, "r--", label = L"W_{0\rightarrow 1}(t)")
    ax1.plot(exc016, "b-.", label = L"W_{0\rightarrow 16}(t)")
    ax1.plot(exc031, "g:", label = L"W_{0\rightarrow 31}(t)")

ax1.set_xlabel(L"\mathrm{Time}\quad [s]")
ax1.set_ylabel(L"\mathrm{Energy\; exchange}\quad W_{i\rightarrow j}(t)")
ax1.set_xticks([0,50,100,150,200])
ax1.set_xticklabels([0,15,30,45,60])

legends = ax1.legend(loc = "upper center", ncol = 3, shadow = true, fontsize = "12", framealpha = 0.75)

tight_layout()

#savefig("circEEx.png", dpi = 300, transparent = true)
```

### 2. Lineal array

```julia
exl01 = wijlt(1, 0, 1, t_list, N, q0);
exl016 = wijlt(1, 0, 16, t_list, N, q0);
exl031 = wijlt(1, 0, 31, t_list, N, q0);
```

```julia
ql0 = [ql(1, 0, t_list[i], N, q0) for i in 1:length(t_list)];
ql1 = [ql(1, 1, t_list[i], N, q0) for i in 1:length(t_list)];
ql16 = [ql(1, 16, t_list[i], N, q0) for i in 1:length(t_list)];
ql31 = [ql(1, 31, t_list[i], N, q0) for i in 1:length(t_list)];
```

```julia
fig_ql = figure(figsize = (10,5))
ax_l = gca()

    ax_l.plot(ql0, "r--", label = L"q_{0}(t)")
    ax_l.plot(ql1, "b-.", label = L"q_{1}(t)")
    ax_l.plot(ql16, "g:", label = L"q_{16}(t)")
    ax_l.plot(ql31, "y-", label = L"q_{31}(t)")

ax_l.set_xlabel(L"\mathrm{Time}\quad [s]")
ax_l.set_ylabel(L"\mathrm{Amplitude}\quad q_{n}(t)")
ax_l.set_xticks([0,50,100,150,200])
ax_l.set_xticklabels([0,15,30,45,60])

ax_l.annotate(L"q_{0}(t)", xy = (1,1))
ax_l.annotate(L"q_{1}(t)", xy = (5,0.5))
ax_l.annotate(L"q_{16}(t)", xy = (52,0.18))
ax_l.annotate(L"q_{31}(t)", xy = (105,0.1))

tight_layout()

legends = ax_l.legend(loc="upper center", ncol = 2, shadow=true, fontsize="12", framealpha = 0.75)

#savefig("linAmp.png", dpi = 300, transparent = true)
```

```julia
fig2 = figure(figsize = (10,5))
ax2 = gca()
    ax2.plot(exl01, "r--", label = L"W_{0\rightarrow 1}(t)")
    ax2.plot(exl016, "b-.", label = L"W_{0\rightarrow 16}(t)")
    ax2.plot(exl031, "g:", label = L"W_{0\rightarrow 31}(t)")

ax2.set_xlabel(L"\mathrm{Time}\quad [s]")
ax2.set_ylabel(L"\mathrm{Energy\; exchange}\quad W_{i\rightarrow j}(t)")
ax2.set_xticks([0,50,100,150,200])
ax2.set_xticklabels([0,15,30,45,60])

axins2 = ax2.inset_axes([0.3, 0.3, 0.4, 0.3])
axins2.plot(exl016, "b-.", label = L"W_{0\rightarrow 16}(t)")
axins2.plot(exl031, "g:", label = L"W_{0\rightarrow 31}(t)")
axins2.set_title(L"W_{0\rightarrow (16,31)}(t)")
axins2.set_xticks([0,50,100,150,200])
axins2.set_xticklabels([0,15,30,45,60])

legends = ax2.legend(loc = "upper center", ncol = 3, shadow = true, fontsize = "12", framealpha = 0.75)

tight_layout()

#savefig("linEEx.png", dpi = 300, transparent = true)
```
