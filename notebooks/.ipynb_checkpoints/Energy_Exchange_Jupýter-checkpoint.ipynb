```julia
using ColabINAOE2021
using PyPlot
using SparseArrays
using LinearAlgebra
using LaTeXStrings
```

```julia
# Global Constants

k = 1
N = 32
;
```

```julia
# Initial amplitude vector constructors

pint(q) = sparsevec(Dict(q + 1 => 1), N + 1); # Single amplitude point at 0 ≤ q ≤ N
```

```julia
# Initial amplitude vector

q0 = normalize(Array(pint(0)));
```

```julia
# Time list

t_list = range(0, stop = 61, length = 200);
```

### 1. Circular array

```julia
ex01 = ColabINAOE2021.wijt(1, 0, 1, t_list, N, q0);
ex016 = ColabINAOE2021.wijt(1, 0, 16, t_list, N, q0);
ex031 = ColabINAOE2021.wijt(1, 0, 31, t_list, N, q0);
```

```julia
fig1 = figure()
ax1 = gca()
    ax1.plot(ex01, "r-", label = L"W_{0\rightarrow 1}")
    ax1.plot(ex016, "b-", label = L"W_{0\rightarrow 16}")
    ax1.plot(ex031, "g-", label = L"W_{0\rightarrow 31}")
ax1.set_xlabel(L"Time\; [s]")
ax1.set_ylabel(L"Energy\; exchange\; W")

legends = ax1.legend(loc="upper center", shadow=true, fontsize="x-large")
```
